```python
import pandas as pd

# Load the uploaded dataset
df = pd.read_csv("ford.csv")

# Display the first 5 rows
df.head()
```

```python
# Separate Independent Features (X) and Dependent Feature (Y)

X = df.drop('price', axis=1)
Y = df['price']

# Print the shape
print("Shape of X:", X.shape)
print("Shape of Y:", Y.shape)
```

```python
# Identify categorical columns
categorical_cols = X.select_dtypes(include=['object']).columns

print("Categorical Columns:")
print(categorical_cols)
```

```python
# Perform One-Hot Encoding
X_encoded = pd.get_dummies(X, columns=categorical_cols)

# Convert all columns to integer type (0 and 1)
X_encoded = X_encoded.astype(int)

# Display the first 5 rows
X_encoded.head()
```

```python
from sklearn.preprocessing import StandardScaler

# List of numerical columns
numerical_cols = ['year', 'mileage', 'tax', 'mpg', 'engineSize']

# Create StandardScaler object
scaler = StandardScaler()

# Apply Standard Scaling
X_encoded[numerical_cols] = scaler.fit_transform(X_encoded[numerical_cols])

# Display the first 5 rows
print("First 5 Rows After Feature Scaling:")
print(X_encoded.head())
```

```python
from sklearn.model_selection import train_test_split

# Split the dataset into training and testing sets
X_train, X_test, y_train, y_test = train_test_split(
    X_encoded,   # Preprocessed features
    Y,           # Target variable
    test_size=0.33,
    random_state=42
)

# Print the shapes
print("Shape of X_train:", X_train.shape)
print("Shape of X_test :", X_test.shape)
print("Shape of y_train:", y_train.shape)
print("Shape of y_test :", y_test.shape)
```

```python
from sklearn.linear_model import LinearRegression

# Create the Linear Regression model
model = LinearRegression()

# Train (fit) the model
model.fit(X_train, y_train)

# Print the intercept
print("Intercept:")
print(model.intercept_)

# Print the coefficients
print("\nCoefficients:")
print(model.coef_)
```

```python
# Make predictions on the test data
y_pred = model.predict(X_test)

# Print the first 10 predicted values
print("First 10 Predicted Values:")
print(y_pred[:10])

# Print the first 10 actual values
print("\nFirst 10 Actual Values:")
print(y_test.iloc[:10].values)
```

```python
from sklearn.metrics import r2_score

# Calculate R2 Score
r2 = r2_score(y_test, y_pred)

# Print R2 Score
print("R2 Score:", r2)
```

```python
import joblib

# Save the Linear Regression model
joblib.dump(model, "LR_ford_car.pkl")

# Save the StandardScaler
joblib.dump(scaler, "scaler.pkl")

# Save the list of columns
columns = X_encoded.columns.tolist()
joblib.dump(columns, "columns.pkl")

print("Model and preprocessing objects saved successfully!")
```
